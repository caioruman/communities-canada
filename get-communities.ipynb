```python
'''
    Read two csv with the location of the first nation and innuit communities in Canada and 
    find the province that they are located using openmaps
'''
import numpy as np
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
# CSV with the communities
fn = 'Premiere_Nation_First_Nation.csv'
innuit = 'Collectivite_Inuite_Inuit_Community.csv'
```

```python
# Oppening the csv
pd1 = pd.read_csv(fn)
pd2 = pd.read_csv(innuit, encoding='ISO-8859-1')
```

```python
geolocator = Nominatim(user_agent="McGill Arctic Communities")
loc_list = []
```

```python
def get_province(lat, lon):
    while True:
        try:
            location = geolocator.reverse("{0}, {1}".format(lat, lon))
            pr = location.raw['address']['state']
            print(lat, lon, pr)
        except:
            print("Time out error on {0}, {1}".format(lat, lon))
            continue
        break
    return pr
```

```python
for lon, lat in zip(pd1['LONGITUDE'], pd1['LATITUDE']):    
    pr = get_province(lat, lon)
    loc_list.append((lat, lon, pr, ''))        
for lon, lat, prov in zip(pd2['LONGITUDE'], pd2['LATITUDE'], pd2['LCO_PROVINCE_CODE']):
    pr = get_province(lat, lon)
    loc_list.append((lat, lon, pr, prov))
```

```python
df = pd.DataFrame(loc_list, columns=['Lat', 'Lon', 'Province', 'Code'])
```

```python
df.to_csv('communities.csv', encoding='ISO-8859-1')
```
